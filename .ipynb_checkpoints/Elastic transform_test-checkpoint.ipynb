```python
from my_utils import *
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# read training data from CSV file 
data = pd.read_csv('input/train.csv')

print('data({0[0]},{0[1]})'.format(data.shape))
print (data.head())
```

```python
## Tranforming image from (0,255) scale to (0,1)

images = data.iloc[:,1:].values
images = images.astype(np.float)

# convert from [0:255] => [0.0:1.0]
images = np.multiply(images, 1.0 / 255.0)

print('images({0[0]},{0[1]})'.format(images.shape))
```

```python
## Calculating image size and image height from number of columns

image_size = images.shape[1]
print ('image_size => {0}'.format(image_size))

# in this case all images are square
image_width = image_height = np.ceil(np.sqrt(image_size)).astype(np.uint8)

print ('image_width => {0}\nimage_height => {1}'.format(image_width,image_height))
```

```python
def display(img):
    
    # (784) => (28,28)
    one_image = img.reshape(image_width,image_height)
    
    plt.axis('off')
    plt.imshow(one_image, cmap=cm.binary)

# output image     
display(images[IMAGE_TO_DISPLAY])
```

```python
## doing elastic transform
import cv2 


## applying elastic erosions and transform
##
## img - should be matrix

def apply_elastic_transform (img):   
    
    img_nes = elastic_transform (img, kernel_dim=17, sigma=6, alpha=9, negated=True)
    return img_nes


def do_all_img_transform (img):
    
    img = img.reshape(image_width,image_height)
    res = apply_elastic_transform(img)
    res = img_xy_scale (res)
    res = crop_and_resize_img (res,image_width,image_height)
    res = img_rotation(res)
    res[res>0.5] = 1 
    
    #kernel = np.zeros((3,3),np.uint8)
    #kernel[1,1] = 1
    #res = cv2.erode(res,kernel,iterations = 1)
    
    res = res.flatten()
    
    return res
```

```python
from sklearn.preprocessing import normalize

figs, axes = plt.subplots(4,2, figsize=(20, 20))


kernel_dim = 17
alpha = 35
sigma = 7
img_proc = images[11].reshape (image_width, image_height) 


# check if the image is a square one
if img_proc.shape[0] != img_proc.shape[1]:
        raise ValueError("Image should be of sqaure form")

# check if kernel dimesnion is odd
if kernel_dim % 2 == 0:
    raise ValueError("Kernel dimension should be odd")

# create an empty image
result = numpy.zeros(img_proc.shape)

# create random displacement fields
displacement_field_x = numpy.array([[random.uniform(-1, 1) for x in xrange(img_proc.shape[0])] \
                            for y in xrange(img_proc.shape[1])]) 
displacement_field_y = numpy.array([[random.uniform(-1, 1) for x in xrange(img_proc.shape[0])] \
                            for y in xrange(img_proc.shape[1])])

#print (displacement_field_x)

axes[0, 0].set_title ("X dist")
axes[0, 0].imshow(displacement_field_x, cmap=cm.binary)
axes[0, 1].set_title ("Y dist")
axes[0, 1].imshow(displacement_field_y, cmap=cm.binary)



# Gaussian Kernel
kern_xy = cv2.getGaussianKernel(kernel_dim, sigma)
displacement_field_x = cv2.sepFilter2D(displacement_field_x, -1, kern_xy,kern_xy)
displacement_field_x = cv2.sepFilter2D(displacement_field_x, -1, kern_xy,kern_xy)
displacement_field_x = cv2.sepFilter2D(displacement_field_x, -1, kern_xy,kern_xy)

displacement_field_y = cv2.sepFilter2D(displacement_field_y, -1, kern_xy,kern_xy)
displacement_field_y = cv2.sepFilter2D(displacement_field_y, -1, kern_xy,kern_xy)
displacement_field_y = cv2.sepFilter2D(displacement_field_y, -1, kern_xy,kern_xy)

## displacement_field_x = cv2.GaussianBlur(displacement_field_x,(sigma,sigma),0)
## displacement_field_y = cv2.GaussianBlur(displacement_field_y,(sigma,sigma),0)

axes[1, 0].set_title ("X dist with Kernel")
axes[1, 0].imshow(displacement_field_x, cmap=cm.binary)
axes[1, 1].set_title ("Y dist with Kernel")
axes[1, 1].imshow(displacement_field_y, cmap=cm.binary)

# Normalization

norm_x = np.sqrt(np.sum(displacement_field_x*displacement_field_x))
norm_y = np.sqrt(np.sum(displacement_field_y*displacement_field_y))
displacement_field_x = displacement_field_x/norm_x * alpha
displacement_field_y = displacement_field_y/norm_y * alpha

print (displacement_field_x.max(),displacement_field_x.min())



# make the distortrd image by averaging each pixel value to the neighbouring
# four pixels based on displacement fields

# bilinear interpolation for:
#
#  [ (low_x,low_y)  (low_x,high_y)
#    (high_x,low_y)  (high_x,high_y)
#

    
for row in xrange(img_proc.shape[1]):
    for col in xrange(img_proc.shape[0]):
            
            #row = 25
            #col = 15
            
            dx = row + displacement_field_x[row, col]
            low_x = int(math.floor(dx))
            high_x = int(math.ceil(dx))

            dy = col + displacement_field_y[row, col]
            low_y = int(math.floor(dy))
            high_y = int(math.ceil(dy))

            #print (dx,dy,low_x,high_x,low_y,high_y)
            
            a_lx_ly = None
            a_lx_hy = None
            a_hx_ly = None
            a_hx_hy = None
            
            if low_x <0 or low_x >= img_proc.shape[0]:
                a_lx_ly = 0
                a_lx_hy = 0
            
            if low_y <0 or low_y>=img_proc.shape[1]:
                a_lx_ly = 0
                a_hx_ly = 0

            if high_x<0 or high_x >= img_proc.shape[0]:
                a_hx_ly = 0
                a_hx_hy = 0
                
            if high_y<0 or high_y >= img_proc.shape[1]:
                a_lx_hy = 0
                a_hx_hy = 0
                
            #print (row,col," ",low_x,high_x,low_y,high_y)
            
            if a_lx_ly is None: a_lx_ly = img_proc[low_x, low_y];        
            if a_lx_hy is None: a_lx_hy = img_proc[low_x, high_y];
            if a_hx_ly is None: a_hx_ly = img_proc[high_x, low_y];
            if a_hx_hy is None: a_hx_hy = img_proc[high_x, high_y]
            
            #print (a_lx_ly, a_lx_hy, a_hx_ly, a_hx_hy)
            
            xy11 = a_lx_ly * (dy - low_y) + a_lx_hy * (high_y - dy)
            xy22 = a_hx_ly * (dy - low_y) + a_hx_hy * (high_y - dy)

            #print (xy11,xy22)
            
            result[row, col] = xy11 * (dx - low_x) + xy22 * (high_x - dx)
            
            #print (result[row, col])
            
            arr_x = displacement_field_x[row, col] / max(displacement_field_x.max(), abs(displacement_field_x.min()))
            arr_y = displacement_field_y[row, col] / max(displacement_field_y.max(), abs(displacement_field_y.min()))
            axes[2, 1].arrow(row, col, arr_x, arr_y,head_width=0.2, fc="k", ec="k") 



# for plotting arrows
zero_img = numpy.zeros(img_proc.shape)
axes[2, 0].imshow(zero_img, cmap=cm.binary)
axes[2, 1].imshow(zero_img, cmap=cm.binary)

    
axes[3, 0].imshow(img_proc, cmap=cm.binary)
axes[3, 1].imshow(result, cmap=cm.binary)
```
